# Task
Generate a full Discounted Cash Flow (DCF) model for a SaaS company in Google Colab. The model should forecast financials for 5 years with the following assumptions: 20% annual revenue growth, 40% COGS, 30% operating expenses, 5% capital expenditures, 4% depreciation, and 2% change in net working capital. Use a 12% discount rate and a 3% terminal growth rate. Compute free cash flows, apply discounting, and calculate enterprise value. Output the results as a structured Pandas DataFrame and export it as an Excel file

Here is all the data you need:
"SaaS_DCF_Model.csv"

## Data loading

### Subtask:
Load the initial financial data from the provided CSV file into a Pandas DataFrame.


**Reasoning**:
Load the provided CSV file into a pandas DataFrame and inspect it.



```python
import pandas as pd

try:
    df = pd.read_csv('SaaS_DCF_Model.csv')
    display(df.head())
    display(df.info())
    display(df.describe())
except FileNotFoundError:
    print("Error: 'SaaS_DCF_Model.csv' not found. Please upload the file to your current working directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data preparation

### Subtask:
Prepare the DataFrame for the DCF model by adding columns for the next 5 years and initializing them.


**Reasoning**:
Create new columns for the next 5 years of financial projections and initialize them to zero, ensuring data type consistency with existing columns.



```python
financial_metrics = ['Revenue ($M)', 'COGS ($M)', 'Operating Expenses ($M)', 'EBITDA ($M)', 'Depreciation ($M)', 'EBIT ($M)', 'Taxes ($M)', 'NOPAT ($M)', 'CapEx ($M)', 'Change in NWC ($M)', 'Free Cash Flow ($M)']

for metric in financial_metrics:
    for i in range(2, 7):  # Years 2 through 6
        new_column_name = f'Year_{i}_{metric.replace(" ", "_").replace("($M)", "")}'
        df[new_column_name] = 0.0
display(df.head())
```

## Data wrangling

### Subtask:
Calculate projected financial metrics for the next 5 years based on the provided growth assumptions.


**Reasoning**:
Calculate projected revenue and related financial metrics for the next 5 years and store them in the corresponding columns of the DataFrame.



```python
# Project revenue for the next 5 years
initial_revenue = df['Revenue ($M)'][0]
for i in range(2, 7):
    year_revenue_column = f'Year_{i}_Revenue_'
    df[year_revenue_column] = initial_revenue * (1 + 0.20)**(i - 1)

    # Calculate other financial metrics based on projected revenue
    cogs_column = f'Year_{i}_COGS_'
    op_exp_column = f'Year_{i}_Operating_Expenses_'
    capex_column = f'Year_{i}_CapEx_'
    depreciation_column = f'Year_{i}_Depreciation_'
    nwc_column = f'Year_{i}_Change_in_NWC_'

    df[cogs_column] = df[year_revenue_column] * 0.40
    df[op_exp_column] = df[year_revenue_column] * 0.30
    df[capex_column] = df[year_revenue_column] * 0.05
    df[depreciation_column] = df[year_revenue_column] * 0.04
    df[nwc_column] = df[year_revenue_column] * 0.02

display(df.head())
```

## Data analysis

### Subtask:
Calculate the Free Cash Flows (FCF) for each year, including the projected years.


**Reasoning**:
Calculate the Free Cash Flow (FCF) for each year using the formula provided and store the results in new columns within the DataFrame.



```python
# Calculate Free Cash Flow for the initial year (Year 1)
df['Free Cash Flow ($M)'] = df['Revenue ($M)'] - df['COGS ($M)'] - df['Operating Expenses ($M)'] - df['CapEx ($M)'] + df['Depreciation ($M)'] - df['Change in NWC ($M)']

# Calculate Free Cash Flow for projected years (Year 2-6)
for i in range(2, 7):
    revenue_col = f'Year_{i}_Revenue_'
    cogs_col = f'Year_{i}_COGS_'
    op_exp_col = f'Year_{i}_Operating_Expenses_'
    capex_col = f'Year_{i}_CapEx_'
    depreciation_col = f'Year_{i}_Depreciation_'
    nwc_col = f'Year_{i}_Change_in_NWC_'
    fcf_col = f'Year_{i}_Free_Cash_Flow_'

    df[fcf_col] = df[revenue_col] - df[cogs_col] - df[op_exp_col] - df[capex_col] + df[depreciation_col] - df[nwc_col]

display(df.head())
```

## Model training

### Subtask:
Discount the projected free cash flows and calculate the terminal value.


**Reasoning**:
Calculate the present value of free cash flows for each projected year (Year 2-6), calculate the terminal value, and calculate the present value of the terminal value.



```python
discount_rate = 0.12
terminal_growth_rate = 0.03

# Calculate the present value of free cash flows for each projected year
for i in range(2, 7):
    fcf_column = f'Year_{i}_Free_Cash_Flow_'
    present_value_column = f'Year_{i}_Present_Value_FCF'
    df[present_value_column] = df[fcf_column] / (1 + discount_rate)**(i - 1)

# Calculate the terminal value
terminal_value = df['Year_6_Free_Cash_Flow_'][0] * (1 + terminal_growth_rate) / (discount_rate - terminal_growth_rate)

# Calculate the present value of the terminal value
present_value_terminal_value = terminal_value / (1 + discount_rate)**5

# Display the results
display(df[['Year_2_Free_Cash_Flow_', 'Year_2_Present_Value_FCF', 'Year_6_Free_Cash_Flow_', 'Year_6_Present_Value_FCF']].head())
print(f"Terminal Value: {terminal_value}")
print(f"Present Value of Terminal Value: {present_value_terminal_value}")
```

## Model evaluation

### Subtask:
Calculate the Enterprise Value by summing the present values of the free cash flows and the present value of the terminal value.


**Reasoning**:
Calculate the sum of present values of free cash flows and add it to the present value of the terminal value to get the enterprise value.



```python
present_value_fcfs = df['Year_2_Present_Value_FCF'].sum() + df['Year_3_Present_Value_FCF'].sum() + df['Year_4_Present_Value_FCF'].sum() + df['Year_5_Present_Value_FCF'].sum() + df['Year_6_Present_Value_FCF'].sum()

# Assuming present_value_terminal_value is already calculated and available
present_value_terminal_value = 43628933.08142962

enterprise_value = present_value_fcfs + present_value_terminal_value

print(f"Enterprise Value: {enterprise_value}")
```
